## Why Python?
计算机语言丰富多彩，从低级语言到高级语言、从解释型语言到编译型语言(**[这是什么？](https://cloud.tencent.com/developer/news/331644)**)，语法规则和学习难度不尽相同。在满足一定条件的情况下，不同的语言可以实现基本相同的任务目标。因此，语言的选择标准可以从如下几个角度进行定义：
- 这门语言的语法结构清晰吗？
- 这门语言的学习资料丰富吗？
- 这门语言是否为问题的解决提供了简洁高效的方案？

Python的语法结构非常清晰，便于直观理解和功能组合；网络上有大量的Python学习资源，初学者遇到的所有问题几乎都能通过浏览器搜索找到答案；Python开源语言的特性使其有大量的库（package,**[有哪些库？](https://pypi.org/)**），灵活使用和修改已封装好的函数可以大大降低编程难度。（因此，很多Python使用者被戏称为*调包侠*。）
<br>
此外，Python可以实现灵活的程序设计理念，包括面向过程（OPP, Process-Oriented Programming）和面向对象(OOP, Object-Oriented Programming)(**[有什么区别？](https://blog.csdn.net/ostrichmyself/article/details/5920427)**)。后续课程中我们会继续研究这一对概念，这对成为一个*调包侠*来说太重要了！

总结一下，Python语言有着以下几个特点：
- Python是通用语言
- Python是脚本语言
- Python是开源语言
- Python是跨平台语言
- Pyhont是多模型语言

## How to learn Python?

学习任何一门语言最重要的是**练习**。在语言学习的起步阶段，对于遇到的每一个代码段，我建议你都亲自把它敲入代码框（而不是使用C/V技术）。亲自运行代码的过程中要注意观察程序的输出与我们的预期是否一致。
- 如果不一致，**一定**去解决它。恭喜你，在解决报错的过程下，你将驶入语言学习的快车道。
- 如果一致，试着换个输入的值、试着换种写法。
我建议你在准备学习Python的过程中解决如下几个问题：
- 安装一个好用的编程环境
- 明确自己使用Python的场景（数据分析？机器学习？计量经济学？或者是其他酷炫的应用场景）
- 学习基础语法知识和编程规范，做一个讨喜的程序员

### 安装Anaconda
Anaconda是一个开源的Python发行版本,提供了包管理与环境管理的功能，可以很方便地解决多版本python并存、切换以及各种第三方包安装问题。其中自带了Jupyter Notebook和Spyder IDE。能够满足各个阶段的用户需求。
<br>
以下的内容以Windows为例，在win10和win11的环境下都可成功安装Anaconda。macOS环境下的安装比较简单，如有疑问可以参考**[安装教程](https://blog.csdn.net/skeey111/article/details/124500951)**。

#### 安装
1. 运行.exe文件，点击Next
2. Select Installation Type: 建议选择Just Me
3. Choose Install Location: 建议在非系统磁盘新建目录下安装
4. Advanced Installation Options: 建议**不勾选**Add Anaconda3 to my PATH environment variable,我们稍后会手动添加；建议**勾选** Register Anaconda3 as my default Python 3.9
5. 点击Install，等待安装完成即可。

#### 路径配置
1. 按下win键，检索“高级系统设置”，点击“环境变量”

2. 双击“系统变量”中的“Path”

3. 点击“新建”，输入红色方框中与anaconda相关的四个路径。注意，此处的环境变量设置要根据Anaconda的安装目录来确定。比如，此处我的Anaconda3文件夹在'D:/Program/'目录下。


#### 检验环境配置是否成功
1. 按下win键，检索“cmd”，打开命令行
2. 输入conda --version并按回车键，若返回conda版本号即为正常
3. 输入python并按回车键，若返回Python版本即为正常
